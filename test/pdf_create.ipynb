```python
if conn:
    try:
        cursor = conn.cursor()
        cursor.execute("""
            SELECT table_name
            FROM information_schema.tables
            WHERE table_schema = 'public' 
            ORDER BY table_name;
        """)
        tables = cursor.fetchall()
        print("Tablas en la base de datos:")
        for table in tables:
            print(f"- {table[0]}")
    except Exception as e:
        print(f"Error al listar tablas: {e}")
    finally:
        if cursor:
            cursor.close()
else:
    print("No hay conexión a la base de datos para listar tablas.")
```

```python
if conn:
    # Reemplaza 'nombre_de_tu_tabla' con el nombre de una de tus tablas
    table_to_view = "clients"

    try:
        df = pd.read_sql_query(f"SELECT * FROM {table_to_view}", conn)
        print(f"\nContenido de la tabla '{table_to_view}':")
        display(df.head()) # Muestra las primeras 5 filas
        print(f"\nNúmero de filas: {len(df)}")
    except Exception as e:
        print(f"Error al leer la tabla '{table_to_view}': {e}")
else:
    print("No hay conexión a la base de datos para leer tablas.")
```
